```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
contest = pd.read_excel("ex1-2/ex1-2/hotdog-contest-winners.xlsm", sheet_name=0, header=0)
sns.barplot(x = 'Country',y = 'Dogs eaten',data = contest)
plt.show()
```

```python
approval = pd.read_excel("ex1-2/ex1-2/obama-approval-ratings.xls", sheet_name=0, header=0)
```

```python
sns.set(style='white')
#create stacked bar chart
approval.set_index('Issue').plot(kind='bar', stacked=True, color=['steelblue', 'darkblue','orange'])
#add overall title
plt.title('Approve ,Disapprove ,None by Issue', fontsize=16)
#add axis titles
plt.xlabel('Issue')
plt.ylabel('Approval Status')
#rotate x-axis labels
plt.xticks(rotation=45)
```

```python
total = contest['Country'].value_counts().values.sum()
def fmt(x):
    return '{:.1f}%\n{:.0f}'.format(x, total*x/100)
plt.pie(contest['Country'].value_counts().values, labels=contest['Country'].value_counts().index, autopct=fmt)
```
